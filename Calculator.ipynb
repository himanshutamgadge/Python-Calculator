```python
last_result = None

def add(num_1, num_2):
  return num_1 + num_2

def sub(num_1, num_2):
  return num_1 - num_2

def mul(num_1, num_2):
  return num_1 * num_2

def div(num_1, num_2):
  if num_2 == 0:
    print("Error: Cannot divide by zero")
    return None
  return num_1 / num_2

def exp(num_1, num_2):
  return num_1 ** num_2

def sqrt(num_1):
  return num_1 ** 0.5

def inp_two_num():
  global last_result
  try:
    use_last = False
    num_1 = None

    if last_result is not None:
      reuse = input(f"Do you want to use the last result ({last_result}) as the first number? (yes/no): ").lower()
      if reuse in ["yes", "y"]:
        num_1 = last_result
        use_last = True

    if not use_last:
      num_1 = float(input("Enter the first number: \n"))

    num_2 = float(input("Enter the second number: \n"))
    return num_1, num_2

  except ValueError:
    print("Error: Please enter a valid number")
    return inp_two_num()

logo='''
   ____      _            _       _
  / ___|__ _| | ___ _   _| | __ _| |_ ___  _ __
 | |   / _` | |/ __| | | | |/ _` | __/ _ \| '__|
 | |__| (_| | | (__| |_| | | (_| | || (_) | |
  \____\__,_|_|\___|\__,_|_|\__,_|\__\___/|_|

   ┌───────────────────┐
   │  7  │  8  │  9  │+│
   │─────┼─────┼─────┼─│
   │  4  │  5  │  6  │-│
   │─────┼─────┼─────┼─│
   │  1  │  2  │  3  │x│
   │─────┼─────┼─────┼─│
   │  0  │  .  │  =  │÷│
   └───────────────────┘

       🧮 PYTHON CALCULATOR 🧮
'''
guide="""
Guide:
Please the number of the operation you want to perform:
01. Addition
02. Subtraction
03. Multiplication
04. Division
05. Exponential
06. Square Root
07. Exit
"""
print(logo)
print(guide)
Stop = False
while not Stop:
  start = input("Please select the operation you want to perform: \n").lower()
  result = None

  if start in ["1", "01", "+"]:
    num_1, num_2 = inp_two_num()
    result = add(num_1, num_2)

  elif start in ["2", "02", "-"]:
    num_1, num_2 = inp_two_num()
    result = sub(num_1, num_2)

  elif start in ["3", "03", "*"]:
    num_1, num_2 = inp_two_num()
    result = mul(num_1, num_2)

  elif start in ["4", "04", "/"]:
    num_1, num_2 = inp_two_num()
    result = div(num_1, num_2)

  elif start in ["5", "05", "**"]:
    try:
      if last_result is not None:
        reuse = input(f"Use last result ({last_result}) as base? (yes/no): ").lower()
        if reuse in ["yes", "y"]:
          num_1 = last_result
        else:
          num_1 = float(input("Enter the number: \n"))
      else:
        num_1 = float(input("Enter the number: \n"))

      num_2 = float(input("Enter the power: \n"))
      result = exp(num_1, num_2)
    except ValueError:
      print("Error: Please enter a valid number")

  elif start in ["6", "06"]:
    try:
      if last_result is not None:
        reuse = input(f"Use last result ({last_result}) for square root? (yes/no): ").lower()
        if reuse in ["yes", "y"]:
          num_1 = last_result
        else:
          num_1 = float(input("Enter the number: \n"))
      else:
        num_1 = float(input("Enter the number: \n"))

      result = sqrt(num_1)
    except ValueError:
      print("Error: Please enter a valid number")

  elif start in ["7", "07"]:
    print("**********Thank you for using the calculator!!**********")
    Stop = True

  else:
    print(f"*****Please enter the correct response*****\n{guide}")
    continue

  if result is not None:
    last_result = result
    print(f"Total: {round(result, 2)}")
```
