```python
def add(num_1,num_2):
  output=num_1+num_2
  print(f"Total: {output}")

def sub(num_1,num_2):
  output=num_1-num_2
  print(f"Total: {output}")

def mul(num_1,num_2):
  output=num_1*num_2
  print(f"Total: {output}")

def div(num_1,num_2):
  if num_2==0:
    print("Error: Cannot divide by zero")
  else:
    output=num_1/num_2
    print(f"Total: {output}")

def exp(num_1,num_2):
  output=num_1**num_2
  print(f"Total: {round(output,2)}")

def sqrt(num_1):
  output=num_1**(1/2)
  print(f"Total: {round(output,2)}")

def inp_two_num():
  try:
    num_1=float(input("Enter the first number: \n"))
    num_2=float(input("Enter the second number: \n"))
  except ValueError:
    print("Error: Please enter a valid number")
    num_1,num_2=inp_two_num()
  return num_1,num_2

logo='''
   ____      _            _       _
  / ___|__ _| | ___ _   _| | __ _| |_ ___  _ __
 | |   / _` | |/ __| | | | |/ _` | __/ _ \| '__|
 | |__| (_| | | (__| |_| | | (_| | || (_) | |
  \____\__,_|_|\___|\__,_|_|\__,_|\__\___/|_|

   â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
   â”‚  7  â”‚  8  â”‚  9  â”‚+â”‚
   â”‚â”€â”€â”€â”€â”€â”¼â”€â”€â”€â”€â”€â”¼â”€â”€â”€â”€â”€â”¼â”€â”‚
   â”‚  4  â”‚  5  â”‚  6  â”‚-â”‚
   â”‚â”€â”€â”€â”€â”€â”¼â”€â”€â”€â”€â”€â”¼â”€â”€â”€â”€â”€â”¼â”€â”‚
   â”‚  1  â”‚  2  â”‚  3  â”‚xâ”‚
   â”‚â”€â”€â”€â”€â”€â”¼â”€â”€â”€â”€â”€â”¼â”€â”€â”€â”€â”€â”¼â”€â”‚
   â”‚  0  â”‚  .  â”‚  =  â”‚Ã·â”‚
   â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜

       ğŸ§® PYTHON CALCULATOR ğŸ§®
'''
guide="""
Guide:
Please the number of the operation you want to perform:
01. Addition
02. Subtraction
03. Multiplication
04. Division
05. Exponential
06. Square Root
07. Exit
"""
print(logo)
print(guide)
Stop=False
while Stop==False:
  start=input("Please select the operation you want to perform: \n").lower()
  if start in ["1","01","+"]: #Addition
    num_1, num_2=inp_two_num()
    add(num_1,num_2)
  elif start in ["2","02","-"]: #Subtraction
    num_1, num_2=inp_two_num()
    sub(num_1,num_2)
  elif start in ["3","03","*"]: #Multiplication
    num_1, num_2=inp_two_num()
    mul(num_1,num_2)
  elif start in ["4","04","/"]:#Division
    num_1, num_2=inp_two_num()
    div(num_1,num_2)
  elif start in ["5","05","**"]:#Exponential
    try:
      num_1=float(input("Enter the number: \n"))
      num_2=float(input("Enter the power: \n"))
      exp(num_1,num_2)
    except ValueError:
      print("Error: Please enter a valid number")
  elif start in ["6","06"]:#Square Root
    try:
      num_1=float(input("Enter the number: \n"))
      sqrt(num_1)
    except ValueError:
      print("Error: Please enter a valid number")
  elif start in ["7","07"]:#Exit
    print("**********Thank you for using the calculator!!**********")
    Stop=True
  else:
    print(f"*****Please enter the correct response*****\n{guide}")
```
