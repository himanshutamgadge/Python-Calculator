```python
menu = {
    "espresso": {
        "ingredients": {
            "water": 50,
            "coffee": 18,
        },
        "cost": 1.5,
    },
    "latte": {
        "ingredients": {
            "water": 200,
            "milk": 150,
            "coffee": 24,
        },
        "cost": 2.5,
    },
    "cappuccino": {
        "ingredients": {
            "water": 250,
            "milk": 100,
            "coffee": 24,
        },
        "cost": 3.0,
    }
}

resources = {
    "water": 300,
    "milk": 200,
    "coffee": 100,
    "money": 0,
}

coins = {
  "quarters": 0.25,
  "dimes": 0.10,
  "nickles": 0.05,
  "pennies": 0.01
}

name_map = {
    "1": "espresso",
    "01": "espresso",
    "espresso": "espresso",
    "2": "latte",
    "02": "latte",
    "latte": "latte",
    "3": "cappuccino",
    "03": "cappuccino",
    "cappuccino": "cappuccino"
}

logo = '''
  /$$$$$$             /$$$$$$   /$$$$$$                          /$$      /$$                     /$$       /$$
 /$$__  $$           /$$__  $$ /$$__  $$                        | $$$    /$$$                    | $$      |__/
| $$  \__/  /$$$$$$ | $$  \__/| $$  \__//$$$$$$   /$$$$$$       | $$$$  /$$$$  /$$$$$$   /$$$$$$$| $$$$$$$  /$$ /$$$$$$$   /$$$$$$
| $$       /$$__  $$| $$$$    | $$$$   /$$__  $$ /$$__  $$      | $$ $$/$$ $$ |____  $$ /$$_____/| $$__  $$| $$| $$__  $$| $$$$$$$$
| $$      | $$  | $$| $$_/    | $$_/  | $$$$$$$$| $$$$$$$$      | $$  $$$| $$  /$$$$$$$| $$      | $$  \ $$| $$| $$  | $$| $$$$$$$$
| $$    $$| $$  | $$| $$      | $$    | $$_____/| $$_____/      | $$\  $ | $$ /$$__  $$| $$      | $$  | $$| $$| $$  | $$| $$_____/
|  $$$$$$/|  $$$$$$/| $$      | $$    |  $$$$$$$|  $$$$$$$      | $$ \/  | $$|  $$$$$$$|  $$$$$$$| $$  | $$| $$| $$  | $$|  $$$$$$$
 \______/  \______/ |__/      |__/     \_______/ \_______/      |__/     |__/ \_______/ \_______/|__/  |__/|__/|__/  |__/ \_______/

'''
MAX_RESOURCES = {
    "water": 1000,
    "milk": 1000,
    "coffee": 1000
}

def welcome():
  print(f'''
              Menu

      Coffee       Price

  01. Espresso     $ {menu["espresso"]["cost"]}
  02. Latte        $ {menu["latte"]["cost"]}
  03. Cappuccino   $ {menu["cappuccino"]["cost"]}

  Type 'admin' to access admin panel
  ''')
  inp=input("\nPlease select your order: ").lower()
  return inp

def money():
  quarters=int(input("How many quarters?: "))
  dimes=int(input("How many dimes?: "))
  nickles=int(input("How many nickles?: "))
  pennies=int(input("How many pennies?: "))
  total=quarters*coins["quarters"]+dimes*coins["dimes"]+nickles*coins["nickles"]+pennies*coins["pennies"]
  return total

def is_enough_resources(drink):
  ingredients = menu[drink]["ingredients"]
  for item in ingredients:
    if resources[item] < ingredients[item]:
      print(f"\nSorry, there is not enough {item}.")
      return False
  return True


def dispense():
  inp = welcome()
  if inp == "admin":
      admin()
      print("*"*180)
      return

  drink_name = name_map.get(inp)

  if not drink_name:
      print("\nInvalid Input")
      print("*"*180)
      return

  if not is_enough_resources(drink_name):
    print("*"*180)
    return

  total = money()

  cost = menu[drink_name]["cost"]
  ingredients = menu[drink_name]["ingredients"]

  if total >= cost:
    change = round(total - cost, 2)
    if change > 0:
      print(f"\nHere is ${change} in change.")
    print(f"\nHere is your {drink_name}. Enjoy!")
    for item in ingredients:
      resources[item] -= ingredients[item]
    resources["money"] += cost
  else:
    print("\nSorry that's not enough money. Money refunded.")
  print("*"*180)


def admin():
  global machine_status
  print('''\n

  01. Print Report
  02. Refil Machine
  03. Turn off the machine
  ''')

  adm = input("\nPlease select from admin panel: ")

  if adm == "1":
    print(f'''\n
    Water: {resources["water"]}ml
    Milk: {resources["milk"]}ml
    Coffee: {resources["coffee"]}g
    Money: ${resources["money"]}
    ''')

  elif adm == "2":
    print("\nRefil Machine")

    for item in ["water", "milk", "coffee"]:
      max_limit = MAX_RESOURCES[item]
      current_amount = resources[item]
      available_capacity = max_limit - current_amount

      print(f"\nCurrent {item.capitalize()}: {current_amount} / {max_limit}")
      if available_capacity == 0:
        print(f"{item.capitalize()} is already full.")
        continue

      try:
        amount = int(input(f"How much {item} do you want to add?: "))
        if amount < 0:
          print("\nCannot add negative amount.")
          continue
        if amount > available_capacity:
          print(f"\n⚠️ Can only add {available_capacity} more. Filling to max.")
          resources[item] = max_limit
        else:
          resources[item] += amount
      except ValueError:
        print("\nInvalid input. Skipping this item.")

    print(f'''
    ✅ Updated Resources:
    Water: {resources["water"]}ml
    Milk: {resources["milk"]}ml
    Coffee: {resources["coffee"]}g
    ''')
    print("*"*180)
  elif adm == "3":
    print("\nMachine turned off")
    print("*"*180)
    machine_status=False


  else:
    print("\nInvalid Input")
    print("*"*180)
    admin()


machine_status=True
print(logo)
while machine_status:
  dispense()
```
